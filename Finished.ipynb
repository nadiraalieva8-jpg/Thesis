```python
# Top-10 words per SDG 


import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_selection import chi2

# NLTK stopwords
import nltk
try:
    from nltk.corpus import stopwords
    _ = stopwords.words("english")
except LookupError:
    nltk.download("stopwords")
from nltk.corpus import stopwords

# -------- CONFIG --------
FILE_PATH  = "/Users/nodiraalieva/Desktop/bachelor/cleaned.xlsx"  # <-- change if needed
TEXT_COL   = "paragraphs"                                         # your text column
LABEL_COLS = ["label_manual", "label_osdg"]                       # run both, separately
TOP_K      = 10                                                   # top words per SDG
MIN_DF     = 2                                                    # ignore words seen in <2 docs (reduce junk)
# ------------------------

# custom stopwords
#aws sdg fable sdgs achievment agency novs star ibm wue ghg reductions led oecd dac gni oda usd nine pathway thank context evolve laws ergs able another accurate
#general want sdgs index subsequent code per employees  sdgs held pride 
#dire sdgs multipe widely gap 
##along trends timely 
custom_stopwords = {
    "amidst", "drove", "invaluable", "convening",
    "make", "use", "section", "company", "project", "report", "aws","sdg","fable","achievment","agency","novs","star","ibm","wue","ghg","reductions","led","oecd","dac","gni","oda","usd","nine","pathway","thank","context", "evolve","laws","ergs","able","another",
    "accurate", "general","want","sds","index","subsequent","code","per","employees","held","pride",
    "dire","sdgs","multiple","widely","gap",
    "along","timely","trends"
}
ALL_STOPWORDS = set(stopwords.words("english")).union(custom_stopwords)

# Load data
df = pd.read_excel(FILE_PATH)
if TEXT_COL not in df.columns:
    raise ValueError(f"Text column '{TEXT_COL}' not found. Available: {list(df.columns)}")

# Basic text clean 
def clean_text(s: str) -> str:
    s = str(s).lower()
    s = re.sub(r'[^a-z\s]', ' ', s)   # keep only letters + spaces
    s = re.sub(r'\s+', ' ', s).strip()
    return s

df[TEXT_COL] = df[TEXT_COL].astype(str).apply(clean_text)

# Vectorize once on the full corpus 
vectorizer = CountVectorizer(
    stop_words=list(ALL_STOPWORDS),
    ngram_range=(1, 1),
    min_df=MIN_DF,
    token_pattern=r"\b[a-z]{3,}\b"   # tokens with >=3 letters
)
X = vectorizer.fit_transform(df[TEXT_COL])
feature_names = np.array(vectorizer.get_feature_names_out())

# normalize SDG labels to numbers 1..17 
def sdg_to_num(val):
    """
    Accepts formats like:
    'sdg1', 'SDG 1', '1', 'SDG 01', '1 - No Poverty', 'SDG13: Climate Action'
    Returns int 1..17 or None if not recognized.
    """
    if pd.isna(val):
        return None
    s = str(val).strip().lower()
    m = re.search(r'\b(?:(?:sdg\s*|sdg-?)?)(1[0-7]|[1-9])\b', s)
    if m:
        n = int(m.group(1))
        if 1 <= n <= 17:
            return n
    return None

# Display helper (works in Jupyter)
def show_table(keywords, scores, title, top_k=10):
    # We show Rank 1..K with highest chi2 at the top
    order_desc = np.argsort(-scores)[:top_k]
    kws = keywords[order_desc]
    scs = scores[order_desc]
    ranks = np.arange(1, len(kws)+1)  # 1..K
    df_out = pd.DataFrame({"Rank": ranks, "Keyword": kws, "Chi2": scs})
    try:
        display(df_out) 
    except NameError:
        print(title)
        print(df_out.to_string(index=False))

# Process each label column independently
for label_col in LABEL_COLS:
    if label_col not in df.columns:
        print(f"Skip: '{label_col}' not found in file. Available: {list(df.columns)}")
        continue

    # Make a normalized numeric SDG column
    norm_col = f"{label_col}__sdgnum"
    df[norm_col] = df[label_col].apply(sdg_to_num)

    print(f"\n==================== {label_col} ====================")
    # For each SDG 1..17, compute top-K one-vs-rest and plot separately
    for sdg_num in range(1, 18):
        # this SDG vs all others 
        y_bin = (df[norm_col] == sdg_num).astype(int).values

        # If no samples for this SDG, skip
        if y_bin.sum() == 0:
            print(f"SDG{sdg_num}: no samples, skipped.")
            continue

        # Chi-square
        chi_scores, _ = chi2(X, y_bin)

        # Pick top-K features, but sort ascending for the horizontal bar 
        top_idx   = np.argsort(chi_scores)[-TOP_K:]
        top_words = feature_names[top_idx]
        top_vals  = chi_scores[top_idx]
        order_asc = np.argsort(top_vals)
        top_words_plot = top_words[order_asc]
        top_vals_plot  = top_vals[order_asc]

        # --- Plot (one SDG at a time) ---
        plt.figure(figsize=(10, 6))
        plt.barh(top_words_plot, top_vals_plot)  
        plt.xlabel("Chi-square score")
        plt.ylabel("Keyword")
        plt.title(f"{label_col} • SDG{sdg_num} • Top {TOP_K} keywords")
        plt.tight_layout()
        plt.show()

        # --- Show table with Rank 1..TOP_K (highest first) ---
        show_table(top_words, top_vals, title=f"{label_col} SDG{sdg_num}", top_k=TOP_K)
#aws sdg fable sdgs achievment agency novs star ibm wue ghg reductions led oecd dac gni oda usd nine pathway thank context evolve laws ergs able another accurate 
#general want sdgs index subsequent code per employees arrears sdgs held pride 
#dire sdgs multipe widey gap 
#along trends timely 
```

```python
# === Overlap (Table 3) + Per-SDG p-values for Manual & OSDG (Table 4 lite) ===
# Requirements (once): pip install pandas scikit-learn scipy nltk openpyxl

import re
import numpy as np
import pandas as pd
from pathlib import Path

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_selection import chi2
from scipy.stats import chi2_contingency

# --- NLTK stopwords (auto-download if missing) ---
import nltk
try:
    from nltk.corpus import stopwords
    _ = stopwords.words("english")
except LookupError:
    nltk.download("stopwords")
from nltk.corpus import stopwords

# -------- CONFIG --------
FILE_PATH  = "/Users/nodiraalieva/Desktop/bachelor/cleaned.xlsx"  # file
TEXT_COL   = "paragraphs"
MANUAL_COL = "label_manual"
OSDG_COL   = "label_osdg"

TOP_K          = 10     # for overlap calc (Table 3)
MIN_DF         = 2      # drop very rare tokens
ALPHA          = 0.05
VOCAB_LIMIT    = 200    # cap vocab size for omnibus p-value calc (keeps fast & stable)
EXPORT_XLSX    = True
EXPORT_PATH    = Path.home() / "Desktop" / "sdg_overlap_and_pvalues.xlsx"
# ------------------------

# ---- Custom stopwords ----
custom_stopwords = {
    "amidst", "drove", "invaluable", "convening",
    "make", "use", "section", "company", "project", "report"
}
ALL_STOPWORDS = set(stopwords.words("english")).union(custom_stopwords)

# --- Helpers ---
def clean_text(s: str) -> str:
    s = str(s).lower()
    s = re.sub(r'[^a-z\s]', ' ', s)   # keep only letters + spaces
    s = re.sub(r'\s+', ' ', s).strip()
    return s

def sdg_to_num(val):
    """
    Parse SDG number from strings like:
    'sdg1', 'SDG 1', '1', 'SDG 01', '1 - No Poverty', 'SDG13: Climate Action'
    Returns int 1..17 or None.
    """
    if pd.isna(val):
        return None
    s = str(val).strip().lower()
    m = re.search(r'\b(?:(?:sdg\s*|sdg-?)?)(1[0-7]|[1-9])\b', s)
    if m:
        n = int(m.group(1))
        if 1 <= n <= 17:
            return n
    return None

# --- Load ---
fp = Path(FILE_PATH)
if not fp.exists():
    raise FileNotFoundError(f"File not found: {fp}")

df = pd.read_excel(fp)
need_cols = [TEXT_COL, MANUAL_COL, OSDG_COL]
missing = [c for c in need_cols if c not in df.columns]
if missing:
    raise ValueError(f"Missing columns: {missing}. Available: {list(df.columns)}")

df = df.dropna(subset=need_cols).copy()
df["text_clean"] = df[TEXT_COL].apply(clean_text)
df["SDG_manual"] = df[MANUAL_COL].apply(sdg_to_num)
df["SDG_osdg"]   = df[OSDG_COL].apply(sdg_to_num)

# --- Vectorize ONCE (stable vocab for all tests) ---
vectorizer = CountVectorizer(
    stop_words=list(ALL_STOPWORDS),
    ngram_range=(1, 1),
    min_df=MIN_DF,
    token_pattern=r"\b[a-z]{3,}\b"
)
X_all = vectorizer.fit_transform(df["text_clean"])
feature_names = np.array(vectorizer.get_feature_names_out())

# --- Top-K chi² keywords per SDG for overlap ---
def topk_keywords_per_sdg(label_col: str, top_k: int):
    out = {}
    labels = df[label_col].astype("Int64")
    for sdg in range(1, 18):
        y = (labels == sdg).fillna(False).astype(int).to_numpy()
        if y.sum() == 0:
            continue
        chi_scores, pvals = chi2(X_all, y)
        order = np.argsort(-chi_scores)[:top_k]
        out[sdg] = list(zip(feature_names[order], chi_scores[order], pvals[order]))
    return out

manual_topk = topk_keywords_per_sdg("SDG_manual", TOP_K)
osdg_topk   = topk_keywords_per_sdg("SDG_osdg",   TOP_K)

# === Table 3: Overlap % Manual vs OSDG (Top-K keywords) ===
rows_overlap = []
for sdg in range(1, 18):
    m_words = {w for (w, _, _) in manual_topk.get(sdg, [])}
    o_words = {w for (w, _, _) in osdg_topk.get(sdg, [])}
    if not m_words and not o_words:
        rows_overlap.append({"SDG": sdg, "Manual_TopK": 0, "OSDG_TopK": 0,
                             "Overlap_Count": 0, "Overlap_%": np.nan})
        continue
    inter = m_words & o_words
    denom = max(len(m_words), len(o_words), 1)
    overlap_pct = round(100.0 * len(inter) / denom, 2)
    rows_overlap.append({
        "SDG": sdg,
        "Manual_TopK": len(m_words),
        "OSDG_TopK": len(o_words),
        "Overlap_Count": len(inter),
        "Overlap_%": overlap_pct
    })

table3 = pd.DataFrame(rows_overlap, columns=["SDG", "Manual_TopK", "OSDG_TopK", "Overlap_Count", "Overlap_%"])

# === Table 4 . p-value per SDG for Manual & OSDG (SDG vs rest) ===
# For each label set separately:
#   - Row 1 = token totals for documents of SDG s
#   - Row 2 = token totals for documents of NOT SDG s
#   - Vocab = top informative terms by chi² vs rest, capped at VOCAB_LIMIT

def omnibus_pvalues_per_label(label_col: str, label_name: str, vocab_limit: int = VOCAB_LIMIT):
    labels = df[label_col].astype("Int64")
    V = len(feature_names)
    results = []
    # Precompute chi² vs rest per SDG to pick informative vocab
    chi2_scores_cache = {}
    for s in range(1, 18):
        y = (labels == s).fillna(False).astype(int).to_numpy()
        if y.sum() == 0:
            chi2_scores_cache[s] = None
            continue
        scores, _ = chi2(X_all, y)
        chi2_scores_cache[s] = scores

    for s in range(1, 18):
        y_pos = (labels == s).fillna(False).to_numpy(dtype=bool)
        n_pos = y_pos.sum()
        y_neg = ~y_pos
        n_neg = y_neg.sum()

        if n_pos == 0:
            results.append({"SDG": s, "Label": label_name, "Docs_Pos": 0, "Docs_Neg": int(n_neg),
                            "Vocab_Size": 0, "Chi2": np.nan, "df": np.nan, "p_value": np.nan,
                            "Decision@0.05": "No data"})
            continue
        if n_neg == 0:
            results.append({"SDG": s, "Label": label_name, "Docs_Pos": int(n_pos), "Docs_Neg": 0,
                            "Vocab_Size": 0, "Chi2": np.nan, "df": np.nan, "p_value": np.nan,
                            "Decision@0.05": "No data"})
            continue

        scores = chi2_scores_cache.get(s)
        if scores is None:
            results.append({"SDG": s, "Label": label_name, "Docs_Pos": int(n_pos), "Docs_Neg": int(n_neg),
                            "Vocab_Size": 0, "Chi2": np.nan, "df": np.nan, "p_value": np.nan,
                            "Decision@0.05": "No data"})
            continue

        top_idx = np.argsort(-scores)[:vocab_limit]
        if top_idx.size == 0:
            results.append({"SDG": s, "Label": label_name, "Docs_Pos": int(n_pos), "Docs_Neg": int(n_neg),
                            "Vocab_Size": 0, "Chi2": np.nan, "df": np.nan, "p_value": np.nan,
                            "Decision@0.05": "No vocab"})
            continue

        # Build 2 x V contingency: SDG vs rest
        X_pos = X_all[y_pos][:, top_idx].sum(axis=0).A1
        X_neg = X_all[y_neg][:, top_idx].sum(axis=0).A1
        if X_pos.sum() == 0 and X_neg.sum() == 0:
            results.append({"SDG": s, "Label": label_name, "Docs_Pos": int(n_pos), "Docs_Neg": int(n_neg),
                            "Vocab_Size": int(top_idx.size), "Chi2": np.nan, "df": np.nan, "p_value": np.nan,
                            "Decision@0.05": "No tokens"})
            continue

        chi2_stat, p_val, dof, _ = chi2_contingency(np.vstack([X_pos, X_neg]), correction=False)
        decision = "Reject H0" if (p_val < ALPHA) else "Fail to reject H0"

        results.append({
            "SDG": s,
            "Label": label_name,
            "Docs_Pos": int(n_pos),
            "Docs_Neg": int(n_neg),
            "Vocab_Size": int(top_idx.size),
            "Chi2": float(chi2_stat),
            "df": int(dof),
            "p_value": float(p_val),
            "Decision@0.05": decision
        })
    return pd.DataFrame(results)

table4_manual = omnibus_pvalues_per_label("SDG_manual", "Manual", vocab_limit=VOCAB_LIMIT)
table4_osdg   = omnibus_pvalues_per_label("SDG_osdg",   "OSDG.ai", vocab_limit=VOCAB_LIMIT)

# Merge into one table (Manual + OSDG rows stacked)
table4 = pd.concat([table4_manual, table4_osdg], ignore_index=True)
table4 = table4[["SDG","Label","Docs_Pos","Docs_Neg","Vocab_Size","Chi2","df","p_value","Decision@0.05"]]

# --- Show ---
print("\n=== Table 3: Overlap % (Manual vs OSDG) per SDG (Top-K keywords) ===")
print(table3.to_string(index=False))

print("\n=== Table 4 (lite): One omnibus p-value per SDG for Manual & OSDG (SDG vs rest) ===")
print(table4.to_string(index=False))

# --- Optional export ---
if EXPORT_XLSX:
    with pd.ExcelWriter(EXPORT_PATH) as writer:
        table3.to_excel(writer, sheet_name="Table3_Overlap", index=False)
        table4.to_excel(writer, sheet_name="Table4_Pvalues", index=False)
    print(f"\nExported: {EXPORT_PATH}")
```
