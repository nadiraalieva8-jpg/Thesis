```python
# === Top-10 words per SDG (multi-label) + p-values + H0 decision ===

import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_selection import chi2

# NLTK stopwords
import nltk
try:
    from nltk.corpus import stopwords
    _ = stopwords.words("english")
except LookupError:
    nltk.download("stopwords")
from nltk.corpus import stopwords

# -------- CONFIG --------
FILE_PATH      = "/Users/nodiraalieva/Desktop/bachelor/cleaned.xlsx"  # <-- change if needed
TEXT_COL       = "paragraphs"                                         # text column
LABEL_COLS     = ["label_manual", "label_osdg"]                       # run both
TOP_K          = 10                                                   # top words per SDG
MIN_DF         = 1                                                    # keep rare words for tiny classes
TOKEN_MINLEN   = 3                                                    # tokens with >=3 letters
NGRAM_RANGE    = (1, 1)                                               # unigrams
ALPHA          = 0.05                                                 # significance level
USE_FDR        = True                                                 # BenjaminiHochberg correction over all features
# ------------------------

# Custom stopwords (extend default)
custom_stopwords = {
    "amidst", "drove", "invaluable", "convening",
    "make", "use", "section", "company", "project", "report",
    "aws","sdg","fable","achievment","agency","novs","star","ibm","wue","ghg","reductions","led","oecd","dac","gni","oda","usd","nine","pathway","thank","context","evolve","laws","ergs","able","another",
    "accurate","general","want","sds","index","subsequent","code","per","employees","held","pride",
    "dire","sdgs","multiple","widely","gap",
    "along","timely","trends"
}
ALL_STOPWORDS = set(stopwords.words("english")).union(custom_stopwords)

# ---------- Load ----------
df = pd.read_excel(FILE_PATH)
if TEXT_COL not in df.columns:
    raise ValueError(f"Text column '{TEXT_COL}' not found. Available: {list(df.columns)}")

# ---------- Clean text ----------
def clean_text(s: str) -> str:
    s = str(s).lower()
    s = re.sub(r'[^a-z\s]', ' ', s)    # keep only letters + spaces
    s = re.sub(r'\s+', ' ', s).strip()
    return s

df[TEXT_COL] = df[TEXT_COL].astype(str).apply(clean_text)

# ---------- Vectorize once on full corpus ----------
vectorizer = CountVectorizer(
    stop_words=list(ALL_STOPWORDS),
    ngram_range=NGRAM_RANGE,
    min_df=MIN_DF,
    token_pattern=rf"\b[a-z]{{{TOKEN_MINLEN},}}\b"
)
X = vectorizer.fit_transform(df[TEXT_COL])
feature_names = np.array(vectorizer.get_feature_names_out())

# ---------- Parse labels as MULTI-LABEL lists ----------
def numbers_in_label(val):
    """
    Returns ALL SDG numbers (1..17) found in a label cell.
    Tolerant: 'Goal 14', 'SDG-14', '14.0', '14, 15', emojis, odd spaces/dashes.
    """
    if pd.isna(val):
        return []
    s = str(val).lower()
    s = s.replace("\u00a0", " ")
    s = re.sub(r"\s+", " ", s)
    s = s.replace("", "-").replace("", "-")
    nums = re.findall(r'(?:sdg|goal)?\s*[-:]?\s*(\d{1,2})', s)
    out = []
    for num in nums:
        try:
            n = int(num)
        except ValueError:
            continue
        if 1 <= n <= 17:
            out.append(n)
    # de-duplicate preserving order
    seen, uniq = set(), []
    for n in out:
        if n not in seen:
            seen.add(n)
            uniq.append(n)
    return uniq

# ---------- BenjaminiHochberg FDR ----------
def benjamini_hochberg(pvals: np.ndarray, alpha: float):
    """
    Returns a boolean mask of which p-values are significant under BH FDR.
    """
    m = pvals.size
    if m == 0:
        return np.array([], dtype=bool)
    order = np.argsort(pvals)
    ranked_p = pvals[order]
    thresholds = alpha * (np.arange(1, m + 1) / m)
    # find largest k where p(k) <= threshold(k)
    passed = ranked_p <= thresholds
    if not passed.any():
        sig_mask = np.zeros_like(pvals, dtype=bool)
    else:
        kmax = np.where(passed)[0].max()
        cutoff = ranked_p[kmax]
        sig_mask = pvals <= cutoff
    return sig_mask

# ---------- Helper: show table ----------
def show_table(df_out, title=None):
    try:
        display(df_out)  # Jupyter
    except NameError:
        if title:
            print(title)
        print(df_out.to_string(index=False))

# ---------- Run per label column ----------
for label_col in LABEL_COLS:
    if label_col not in df.columns:
        print(f"Skip: '{label_col}' not found in file. Available: {list(df.columns)}")
        continue

    print(f"\n==================== {label_col} ====================")

    # Parse to lists of SDG numbers
    parsed_lists = df[label_col].apply(numbers_in_label)

    # Diagnostics: count rows that contain each SDG at least once
    counts = {k: int(parsed_lists.apply(lambda lst, k=k: k in lst).sum()) for k in range(1, 18)}
    print("Rows containing each SDG (multi-label aware):")
    print({k: v for k, v in counts.items() if v > 0})
    missing = [k for k, v in counts.items() if v == 0]
    if missing:
        print("No rows contain SDGs:", missing)

    # Per SDG: one-vs-rest chi-square with p-values and H0 decision
    sdg_summaries = []

    for sdg_num in range(1, 18):
        y_bin = parsed_lists.apply(lambda lst, s=sdg_num: s in lst).astype(int).values
        n_pos = int(y_bin.sum())
        if n_pos == 0:
            print(f"SDG{sdg_num}: no samples, skipped.")
            continue

        # Chi2 + p-values for each feature
        chi_scores, p_values = chi2(X, y_bin)

        # Optional FDR across all features for this SDG
        if USE_FDR:
            sig_mask = benjamini_hochberg(p_values, ALPHA)
            alpha_used = f"FDR-BH {ALPHA}"
        else:
            sig_mask = p_values < ALPHA
            alpha_used = f"{ALPHA}"

        n_sig = int(sig_mask.sum())
        min_p = float(p_values.min()) if p_values.size else float("nan")
        decision = "Reject H0" if n_sig > 0 else "Fail to reject H0"

        sdg_summaries.append({
            "SDG": sdg_num,
            "n_pos": n_pos,
            "Significant_Features": n_sig,
            "Min_p_value": min_p,
            "Alpha/Criterion": alpha_used,
            "Decision": decision
        })

        # ----- TOP-K TABLE (with p-values) -----
        order_desc = np.argsort(-chi_scores)
        top_idx = order_desc[:TOP_K]
        df_top = pd.DataFrame({
            "Keyword": feature_names[top_idx],
            "Chi2": chi_scores[top_idx],
            "p_value": p_values[top_idx],
            "Significant": sig_mask[top_idx]
        }).sort_values("Chi2", ascending=False).reset_index(drop=True)

        # Plot (Top-K)
        plt.figure(figsize=(10, 6))
        # sort ascending for nicer horizontal bars
        plot_order = np.argsort(df_top["Chi2"].values)
        plt.barh(
            df_top["Keyword"].values[plot_order],
            df_top["Chi2"].values[plot_order]
        )
        plt.xlabel("Chi-square score")
        plt.ylabel("Keyword")
        plt.title(f"{label_col}  SDG{sdg_num}  Top {TOP_K} keywords (n={n_pos})\n"
                  f"{decision} @ {alpha_used} | sig features: {n_sig} | min p={min_p:.2e}")
        plt.tight_layout()
        plt.show()

        # Show table
        print(f"Top {TOP_K} keywords for {label_col} SDG{sdg_num} (alpha={alpha_used})")
        show_table(df_top)

    # ----- Summary per SDG for this label_col -----
    if sdg_summaries:
        df_summary = pd.DataFrame(sdg_summaries).sort_values("SDG")
        print(f"\n=== Summary: {label_col} (alpha={alpha_used}) ===")
        show_table(df_summary, title=f"Summary {label_col}")
```
